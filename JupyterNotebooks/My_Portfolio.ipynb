```python
from finance_python.stock import stock
from finance_python.industry import industry
import sys, os, time
import numpy as np
import pandas
from datetime import datetime, date, timedelta
from pandas import DataFrame
```

```python
symbols = ['PCM', 'AHC', 'VEDL', 'DFP', 'PCI', 'MO', 'BTZ', 'WHF', 'ABR', 'JCAP',
              'BANX', 'HRZN', 'ETJ', 'HYI', 'IRM', 'RBS', 'LMRK', 'WHG', 'CCD', 'GSBD', 'CHY', 
                 'PMD']

symbols = set(symbols)
symbols = list(symbols)
start = date.today() - timedelta(days=365)
end = date.today()
```

```python
stock_loop = lambda x: stock(x, start, end)
stat = lambda x: stocks[x].stats()
bal = lambda x: stocks[x].balance()
fin = lambda x: stocks[x].financial()
ca = lambda x: stocks[x].cash()
an = lambda x: stocks[x].analyze()
```

```python
timer_start = time.time()

stocks = list(map(stock_loop, symbols))

timer_end = time.time()
print(timer_end - timer_start)
```

```python
timer_start = time.time()

for each in range(0, len(stocks)):
    stat(each)
    
timer_end = time.time()
print(timer_end - timer_start)
```

```python
timer_start = time.time()

for each in range(0, len(stocks)):
    bal(each)
    
timer_end = time.time()
print(timer_end - timer_start)
```

```python
timer_start = time.time()

for each in range(0, len(stocks)):
    fin(each)
    
timer_end = time.time()
print(timer_end - timer_start)
```

```python
timer_start = time.time()

for each in range(0, len(stocks)):
    ca(each)
    
timer_end = time.time()
print(timer_end - timer_start)
```

```python
timer_start = time.time()

for each in range(0, len(stocks)):
    an(each)
    
timer_end = time.time()
print(timer_end - timer_start)
```

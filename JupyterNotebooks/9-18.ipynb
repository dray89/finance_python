```python
symbols = ['WEED', 'BHC', 'ACB', 'CRON', 'APHA', 'HEXO', 'GUD', 'CWEB', 'OGI', 'TGOD', 'LABS', 'NEPT', 'FIRE', 'HLS', 'RIV', 'TRST', 'ALEF', 'CORV']

start = date.today() - timedelta(days=365)
end = date.today()
```

```python
stock_loop = lambda x: stock(x+'{0}'.format('.TO'), start, end)
stat = lambda x: stocks[x].stats()
bal = lambda x: stocks[x].balance()
fin = lambda x: stocks[x].financial()
ca = lambda x: stocks[x].cash()
an = lambda x: stocks[x].analyze()
```

```python
timer_start = time.time()

stocks = list(map(stock_loop, symbols))

timer_end = time.time()
print(timer_end - timer_start)
```

```python
timer_start = time.time()

for each in range(0, len(stocks)):
    stat(each)
    
timer_end = time.time()
print(timer_end - timer_start)
```

```python
timer_start = time.time()

for each in range(0, len(stocks)):
    bal(each)
    
timer_end = time.time()
print(timer_end - timer_start)
```

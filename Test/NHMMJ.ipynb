```python
search = ['Namaste-technologies-horizon-marijuana']
```

```python
start_time = time.time()
Weed = News(search[0])
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
len(Weed.links)
```

```python
arts = Weed.art_list

for each in arts:
    if each.summary == ' ':
        print(each.index)
    else:
        pass
```

```python
len(arts)
```

```python
for each in arts:
    print(each.title, each.authors, '\n', each.summary, '\n' + '-'*100)
```

```python
keywords = set()
for each in arts:
    for i in each.keywords:
        keywords.add(i)
```

```python
for each in keywords:
    print(each + ",")
```

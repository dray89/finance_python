```python
symbols = ['EMA.TO', 'KEG-UN.TO', 'IGM.TO', 'SHOP.TO', 
           'TA.TO' ,'BX','BEP-UN.TO', 'TRI.TO']

start = date.today() - timedelta(days=365)
end = date.today()
```

```python
stock_loop = lambda x: stock(x, start, end)
stat = lambda x: stocks[x].stats()
bal = lambda x: stocks[x].balance()
fin = lambda x: stocks[x].financial()
ca = lambda x: stocks[x].cash()
an = lambda x: stocks[x].analyze()
```
